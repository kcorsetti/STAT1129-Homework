```python
L=[]
f = open("numbers.txt", "r")
for x in f:
    x = x.split('\n')[0]
    numbers=x.split(',')
    L = L + numbers
L
```

```python
# L = open("numbers.txt", "r")
d = dict()
  
for number in L:
        if number in d:
            d[number] = d[number] + 1
        else:
            d[number] = 1

for key in list(d.keys()):
    print(key, ":", d[key])
```

```python
#Make histogram of frequencies
import numpy as np
import matplotlib.pyplot as plt

plt.bar(d.keys(), d.values())
```

```python
#Convert to json file
import json
y=json.dumps(d)

print(y)
```

```python
#Store into local computer
save_file = open("savedata.json", "w")  
json.dump(d, save_file)  
save_file.close()  
```

```python
#Problem 2
import numpy as np
import pandas as pd

#load csv file
df = pd.read_csv('amazon-orders.csv')

df=df.replace(np.nan,0,regex=True)

print(df)
```

```python
#Getting descriptive statistics of "Total Charged"
df["Total Charged"]
df["Total Charged value"] = [item[1:] for item in df["Total Charged"]]
df["Total Charged value"] = df["Total Charged value"].astype(float)
TC = df["Total Charged value"].astype(float)
TC.describe()
```

```python
#Getting descriptive statistics for each month
april = df.loc[0:10, "Total Charged value"]
april.describe()
```

```python
may = df.loc[10:28, "Total Charged value"]
may.describe()
```

```python
june = df.loc[28:34, "Total Charged value"]
june.describe()
```

```python
july = df.loc[34:, "Total Charged value"]
july.describe()
```

```python
#Bar chart to analyze total charges by month
import matplotlib.pyplot as plt

plt.bar(['April','May','June', 'July'],[sum(april),sum(may),sum(june), sum(july)])
```

```python
#Pie chart to analyze total charges by month
import numpy as np

y = np.array([412.15, 323, 142.275, 954.13])
mylabels = ["April", "May", "June", "July"]

plt.pie(y, labels = mylabels)
plt.show()
```
