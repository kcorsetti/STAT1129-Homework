```python
#Problem 5

x = np.array(["Andy", "Amy", "James", "Jules", "Arthur"])
y = np.array([88, 66, 90, 55, 77])

plt.bar(x, y, width=.5)
plt.title("Student Grades")
plt.show()
```

```python
mylabels = ["Andy", "Amy", "James", "Jules", "Arthur"]
plt.pie(y, labels=mylabels)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

```python
#Problem 6
x = np.array([3, 6, 9, 12])
y = np.array([2,8, 1, 10])
plt.subplot(2, 3, 1)
plt.plot(x, y)
plt.title("Subplot Practice 1")

x = np.array([4, 5, 7, 9])
y = np.array([2, 5, 8, 10])
plt.subplot(2, 3, 2)
plt.plot(x, y)
plt.title("Subplot Practice 2")
plt.show()
```

```python
#Multiple Lines
y1 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([5, 6, 7, 8, 9, 10])
y3 = np.array([9, 10, 11, 12, 13])
plt.subplot(2, 3, 1)
plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
plt.title("Subplot Practice 1")

y1 = np.array([4, 5, 7, 9])
y2 = np.array([2, 5, 8, 10])
y3 = np.array([9, 10, 11, 12])

plt.subplot(2, 3, 2)
plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
plt.title("Subplot Practice 2")
plt.show()
```

```python
#bar chart
y1 = np.array([0, 1, 2, 3])
y2 = np.array([5, 6, 7, 8])

plt.subplot(2, 3, 1)
plt.bar(y1, y2)
plt.title("Subplot Practice 1")

y1 = np.array([4, 5, 7, 9])
y2 = np.array([2, 5, 8, 10])

plt.subplot(2, 3, 2)
plt.bar(y1, y2)
plt.title("Subplot Practice 2")
plt.show()
```

```python
#pie chart
y1 = np.array([0, 1, 2, 3])

plt.subplot(2, 3, 1)
plt.pie(y1)
plt.title("Subplot Practice 1")

y1 = np.array([4, 5, 7, 9])

plt.subplot(2, 3, 2)
plt.pie(y1)
plt.title("Subplot Practice 2")
plt.show()
```

```python
#With Grid

x = np.array([3, 6, 9, 12])
y = np.array([2,8, 1, 10])
plt.subplot(2, 3, 1)
plt.plot(x, y)
plt.grid()
plt.title("Subplot Practice 1")

x = np.array([4, 5, 7, 9])
y = np.array([2, 5, 8, 10])
plt.subplot(2, 3, 2)
plt.plot(x, y)
plt.grid()
plt.title("Subplot Practice 2")
plt.show()
```

```python
#scatter plot
y1 = np.array([0, 1, 2, 3])
y2 = np.array([5, 6, 7, 8])

plt.subplot(2, 3, 1)
plt.scatter(y1, y2)
plt.title("Subplot Practice 1")

y1 = np.array([4, 5, 7, 9])
y2 = np.array([2, 5, 8, 10])

plt.subplot(2, 3, 2)
plt.scatter(y1, y2)
plt.title("Subplot Practice 2")
plt.show()
```

```python
#Histogram
mu, sigma = 0, .1
s = np.random.normal(mu, sigma, 100)
plt.subplot(2, 3, 1)
plt.hist(s)
plt.title("Subplot Practice 1")
plt.show()

mu, sigma = 0, .4
s = np.random.normal(mu, sigma, 100)
plt.subplot(2, 3, 2)
plt.hist(s)
plt.title("Subplot Practice 2")
plt.show()
```
